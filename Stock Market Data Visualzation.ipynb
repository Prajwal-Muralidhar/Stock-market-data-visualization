```python
#libraries for importing data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from pandas_datareader import data,wb
%matplotlib inline 
```

```python
#start and end date of stocks and tickers
start_date= '2000-01-01'
end_date= '2022-12-30'

ticks =['MSFT','AMZN','AAPL','GOOG','META']
```

```python
#downloading stock data using yfinance
df=yf.download(ticks,start_date,end_date)
df
```

```python
#first five values
df.head()
```

```python
MSFT= yf.download('MSFT',start_date,end_date)
AAPL= yf.download('AAPL',start_date,end_date)
GOOG= yf.download('GOOG',start_date,end_date)
META= yf.download('META',start_date,end_date)
AMZN= yf.download('AMZN',start_date,end_date)
```

```python
#adjusted close prices of stocks
yf.download(ticks,start_date,end_date)["Adj Close"]
```

```python
adj=df['Adj Close']
```

```python
#plottting of Adj close price
adj.plot()
plt.title('Adjusted Close Price')
plt.xlabel("Year");
plt.ylabel("Price")

plt.show()
```

```python
#calculating simple return
daily_return= adj.pct_change()
```

```python
#daily return of stocks
daily_return
```

```python
#first five values
daily_return.head()
```

```python
#plotting daily return
daily_return.plot()
plt.title('Daily return of stocks')
plt.xlabel("Year");
plt.ylabel("Price")

plt.show()
```

```python
#creating subplot and plotting 
daily_return.plot(title = 'Daily returns of stocks', subplots = True, figsize=(10,10))
```

```python
#calculating mean rates of daily return
dr_mean= daily_return.mean()
```

```python
#mean values of daily return
dr_mean
```

```python
#plotting using matplotlib
#creating data set
data= {'AAPL':0.001198, 'AMZN':0.001008, 'GOOG':0.000958, 'META':0.000751, 'MSFT':0.000513}
ticks = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (8, 4))
#creating bar plot
plt.bar(ticks, values, color ='maroon', width = 0.4)
plt.xlabel("Stocks")
plt.ylabel("Values")
plt.title("Mean of daily return")
plt.show()
```
